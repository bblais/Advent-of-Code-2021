```python
text="""5483143223
2745854711
5264556173
6141336146
6357385478
4167524645
2176841721
6882881134
4846848554
5283751526
""".strip()

def text2arr(text):
    lines=text.strip().split('\n')
    arr_str=[[int(_) for _ in line] for line in lines]
    arr=array(arr_str)
    return arr

arr=text2arr(text)
arr
```

```python
# @njit
def step(arr):
    arr=arr.copy()
    r,c=arr.shape
    
    for i in range(r):
        for j in range(c):
            arr[i,j]+=1
            
    flash=True
    while flash:
        flash=False
        for i in range(r):
            for j in range(c):
                if arr[i,j]>9:
                    arr[i,j]=0  # flash!
                    flash=True

                    for i2 in [i-1,i,i+1]:
                        for j2 in [j-1,j,j+1]:
                            if i2<0 or j2<0 or i2>=r or j2>=c or (i2==i and j2==j):
                                continue
                            if arr[i2,j2]>0:
                                arr[i2,j2]+=1

    return arr
```
