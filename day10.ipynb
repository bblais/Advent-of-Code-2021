```python
text="""
[({(<(())[]>[[{[]{<()<>>
[(()[<>])]({[<{<<[]>>(
{([(<{}[<>[]}>{[]{[(<()>
(((({<>}<{<{<>}{[]{[]{}
[[<[([]))<([[{}[[()]]]
[{[{({}]{}}([{[{{{}}([]
{<[[]]>}<{[{[{[]{()[[[]
[<(<(<(<{}))><([]([]()
<{([([[(<>()){}]>(<<{{
<{([{{}}[<[[[<>{}]]]>[]]
""".strip()
lines=text.strip().split('\n')
```

```python
scores={")":3,"]":57,"}":1197,">":25137}
```

```python
score=0
for line in lines:

    next_c=[]
    error=False
    for i,c in enumerate(line):
        if c=='[':
            next_c.append(']')
        elif c=='(':
            next_c.append(')')
        elif c=='{':
            next_c.append('}')
        elif c=='<':
            next_c.append('>')
        else:
            nc=next_c.pop()
            if nc!=c:
                error=True
                break

    if error:
        score+=scores[c]
        print(f"Error {c} at {i} for '{line}' {scores[c]} score {score}")
```
